```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix, classification_report, ConfusionMatrixDisplay
import matplotlib.pyplot as plt
```

```python
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.2, random_state=42)   #splitting the dataset
print(x_train.shape, x_test.shape, y_train.shape, y_test.shape)
```

```python
y.value_counts()
```

```python
weight_0=6665/(2*6446)   # calculating weightage of classes
weight_1=6665/(2*219)
```

```python
rf_all=RandomForestClassifier(n_estimators=200, random_state=42,class_weight={0:weight_0,1:weight_1})  # embedded feature selection approach
rf_all.fit(x_train,y_train)
y_pred_all = rf_all.predict(x_test)
cm_all = confusion_matrix(y_test, y_pred_all)
print("Confusion Matrix (All Features):\n", cm_all)

disp_all = ConfusionMatrixDisplay(confusion_matrix=cm_all, display_labels=rf_all.classes_)
disp_all.plot(cmap='Blues')
plt.title("Confusion Matrix - All Features")
plt.show()
```

```python
importances=rf_all.feature_importances_   # feature selection
feature_names=x.columns
importance_df=pd.DataFrame({'Feature':feature_names, 'Importance':importances})
top_30_features=importance_df.sort_values(by='Importance',ascending=False).head(30)
top_30_features
```

```python
x_train_20=x_train[top_30_features['Feature']]   # filtering features from teat and train data
x_test_20=x_test[top_30_features['Feature']]
```

```python
print(x_train_20.shape, x_test_20.shape, y_train.shape, y_test.shape)
```

```python
x_test_20
```

```python
rf=RandomForestClassifier(n_estimators=80,random_state=42,class_weight={0:weight_0,1:90})  # random forest algorithm trained by inducing class weightage in parameters rather than sampling
rf.fit(x_train_20,y_train)
y_pred20=rf.predict(x_test_20)
cm_20=confusion_matrix(y_test,y_pred20)
disp_20 = ConfusionMatrixDisplay(confusion_matrix=cm_20, display_labels=rf.classes_)
disp_20.plot(cmap='Blues')
plt.title("Confusion Matrix - Top 20 Features")
plt.show()
```

```python
x_resam,y_resam=sm.fit_resample(x_train_20,y_train)    #SMOTE resampling
print(x_resam.shape,y_resam.shape)
```

```python
y_resam.value_counts()
```

```python
# training and testing Random forest algorithm with confusion matrix

rf_resam=RandomForestClassifier(n_estimators=80,random_state=42,criterion='gini')
rf_resam.fit(x_resam,y_resam)

y_proba = rf_resam.predict_proba(x_test_20)[:, 1]
y_pred_custom = (y_proba >= 0.27).astype(int)
cm_20=confusion_matrix(y_test,y_pred_custom)
disp_20 = ConfusionMatrixDisplay(confusion_matrix=cm_20, display_labels=rf_resam.classes_)
disp_20.plot(cmap='Blues')
plt.title("Confusion Matrix - Random Forest")
plt.show()
```

```python
rf_score=rf_resam.score(x_resam,y_resam)
rf_score
```

```python
fractions = np.linspace(0.1, 1.0, 10)  #Generating and plotting learning curve
train_scores = []
test_scores  = []
for frac in fractions:
    n = int(frac * len(x_resam))
    idx = np.random.RandomState(42).choice(len(x_resam), size=n, replace=False)
    X_sub, y_sub = x_resam.iloc[idx], y_resam.iloc[idx]
    rf_resam.fit(X_sub, y_sub)
    train_scores.append(rf_resam.score(X_sub, y_sub))
    test_scores.append(rf_resam.score(x_test_20, y_test))

# Plot
plt.figure(figsize=(8, 5))
plt.plot(fractions*100, train_scores, label='Train Accuracy')
plt.plot(fractions*100, test_scores,  label='Test Accuracy')
plt.fill_between(fractions*100,
                 np.array(train_scores) - np.std(train_scores),
                 np.array(train_scores) + np.std(train_scores),
                 alpha=0.1)
plt.fill_between(fractions*100,
                 np.array(test_scores) - np.std(test_scores),
                 np.array(test_scores) + np.std(test_scores),
                 alpha=0.1)
plt.xlabel('Percentage of Resampled Train Data Used (%)')
plt.ylabel('Accuracy')
plt.title('Learning Curve')
plt.legend(loc='best')
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import roc_curve, roc_auc_score  # Generating and plotting ROC curve
y_proba = rf_resam.predict_proba(x_test_20)[:, 1]


fpr, tpr, thresholds = roc_curve(y_test, y_proba)

auc_score = roc_auc_score(y_test, y_proba)

# 4. Plot
plt.figure(figsize=(6,6))
plt.plot(fpr, tpr, lw=2, label=f'ROC Curve (AUC = {auc_score:.2f})')
plt.plot([0,1], [0,1], linestyle='--', color='gray', label='Random Guess')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve — Random Forest (resampled)')
plt.legend(loc='lower right')
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```
